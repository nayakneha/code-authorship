```python
import os
import json

import numpy as np

from matplotlib import pyplot as plt
```

```python
def read_feature_result(path):
    def func():
        with open(path) as f:
            for line in f:
                if not line.startswith('JSON-RESULT='):
                    continue
                data = json.loads(line.split('=', 1)[1])
                yield data
    records = list(func())
    return records
```

```python
token_types = [
'comment',
'string',
'newline',
'number',
'indent',
'dedent',
'encoding',
'endmarker',
'errortoken',
'nl',
'name',
'op',
]
```

```python
path_nofeatures = '/Users/adrozdov/School/amir-security/code-authorship/experiments/results/nofeature.txt'
path_onlyfeatures = '/Users/adrozdov/School/amir-security/code-authorship/experiments/results/onlyfeature.txt'
```

# Ablation (All features except 1 or 2)

```python
records = read_feature_result(path_nofeatures)
```

```python
# Build no-grid.

no_grid = {}

for i1, tt1 in enumerate(token_types):
    for i2, tt2 in enumerate(token_types):
        
        key = (tt1, tt2)
        
        # Get relevant records.
        relevant_records = []
        for x in records:
            keep = True
            
            # Ignore other (tt1, tt2) instances.
            for tt3 in token_types:
                flagname = 'no{}'.format(tt3)
                flagval = True if tt3 in (tt1, tt2) else False
                if x['flags'][flagname] != flagval:
                    keep = False
                    break
                    
            # Ignore `only` instances.
            for tt3 in token_types:
                flagname = 'only{}'.format(tt3)
                flagval = False
                if x['flags'][flagname] != flagval:
                    keep = False
                    break 
            
            if keep:
                relevant_records.append(x)
                
        # There should be at most 2 of each type of experiment. Keep the first.
        
        if len(relevant_records) == 2:
            relevant_records = relevant_records[:1]

        # Assert.
        
        if len(relevant_records) != 1:
            for x in relevant_records:
                print(x)
            raise Exception('Did not find correct number of records. length={}, tt1={}, tt2={}'.format(
                len(relevant_records), tt1, tt2))
            
        # Save to the grid.
        
        no_grid[key] = relevant_records[0]
```

```python
N = len(token_types)

accuracy = np.zeros((N, N))

for i1, tt1 in enumerate(token_types):
    for i2, tt2 in enumerate(token_types):
        if i1 >= i2:
            accuracy[i1, i2] = round(no_grid[(tt1, tt2)]['average_acc'], 3)

fig = plt.figure(figsize=(8, 8))

# Heatmap.
plt.imshow(accuracy)

# Accuracy.
for i in range(N):
    for j in range(N):
        if i >= j:
            text = plt.text(j, i, accuracy[i, j],
                           ha="center", va="center", color="w")

        
# ---- Aesthetics ---- #

ax = fig.get_axes()[0]

# We want to show all ticks...
ax.set_xticks(np.arange(N))
ax.set_yticks(np.arange(N))
# ... and label them with the respective list entries
ax.set_xticklabels(token_types)
ax.set_yticklabels(token_types)
        
# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

plt.title('Feature Ablation (Keep all except 1 or 2 features)')

plt.tight_layout()

plt.show()
```

# Minimum Author Threshold (Reserved vs. User-Defined)

```python
def readfile(path):
    def func():
        with open(path) as f:
            for line in f:
                if not line.startswith('JSON-RESULT='):
                    continue
                data = json.loads(line.split('=', 1)[1])
                yield data
    records = list(func())
    return records

path = './experiments/results/minthreshold.txt'
records = readfile(path)

plt.figure()

relevant = [x for x in records if x['flags']['name'].startswith('onlyreserved')]
xs = [x['flags']['minthreshold_author'] for x in relevant]
ys = [x['average_acc'] for x in relevant]
print(len(ys))

plt.plot(xs, ys, label='Reserved')

relevant = [x for x in records if x['flags']['name'].startswith('noreserved')]
xs = [x['flags']['minthreshold_author'] for x in relevant]
ys = [x['average_acc'] for x in relevant]
print(len(ys))

plt.plot(xs, ys, label='User-Defined')

relevant = [x for x in records if x['flags']['name'].startswith('both')]
xs = [x['flags']['minthreshold_author'] for x in relevant]
ys = [x['average_acc'] for x in relevant]
print(len(ys))

plt.plot(xs, ys, label='All NAME')

plt.legend(loc='best')
plt.yticks(np.arange(10)/10)
plt.title('Performance by token usage')
plt.ylabel('Average accuracy')
plt.xlabel('Minimum token usage (among authors)')
plt.tight_layout()
plt.show()
```
